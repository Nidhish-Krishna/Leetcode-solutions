# Given an array of integers, return indices of the two numbers such that they add up to a specific target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

Given nums = [2, 7, 11, 15], target = 9,

Because nums[0] + nums[1] = 2 + 7 = 9,
return [0, 1].

# Method 1: Brute force

```python
def twosum(nums,t):
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if nums[j]==t-nums[i]:
                return ([i,j])
nums=[2,7,11,15]
t=9
print(twosum(nums,t))
```

# Method 2: Simple solution

```python
def twosum(nums,t):
    for i in range(len(nums)):
        c=t-nums[i]
        if c in nums:
            p=nums.index(c)
            if p!=i:
                return [i,p]
nums=[2,7,11,15]
t=9
print(twosum(nums,t))
```

# Method 3: Two pass hashing

```python
def twosum(nums,t):
    d={}
    for ele in nums:
        d[ele]=nums.index(ele)
    for i in range(len(nums)):
        c=t-nums[i]
        if c in d and d[c]!=i:
            return [i,d[c]]
        
nums=[2,7,11,15]
t=9
print(twosum(nums,t))
```

# Method 4: One pass hashing

```python
def twosum(nums,t):
    d={}
    for i in range(len(nums)):
        c=t-nums[i]
        if c in d:
            return [d[c],i]
        d[nums[i]]=i
        
nums=[2,7,11,15]
t=26
print(twosum(nums,t))
```
